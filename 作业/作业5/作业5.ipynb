（1）读取数据。读取之前作业保存的“酒店数据1.xlsx”

```python
hotel = pd.read_excel("酒店数据1.xlsx")
hotel = hotel.drop("Unnamed: 0",axis=1)
pd.set_option('max_columns',100)  ##设置显示的最大列
pd.set_option('max_rows',500)  ##设置显示的最大行
```

（2）将“类型”和“名字”设置为层次化索引，并交换索引的位置。然后将层次化索引取消。

```python
hotel = hotel.set_index(["类型","名字"])
hotel
```

```python
hotel=hotel.swaplevel("类型","名字")
```

```python
hotel=hotel.reset_index()
```

（4）用Groupby方法来计算每个地区的评分人数的总和以及均值。

```python
hotel["评分人数"].groupby(hotel["地区"]).sum()
```

（5）用Grouby方法计算每个类型的平均价格，最高价和最低价。

```python
hotel["价格"].groupby(hotel["类型"]).mean()
```

```python
hotel["价格"].groupby(hotel["类型"]).max()
```

（6）数据离散化，按照价格将酒店分为3个等级,0-500为C，500-1000为B,大于1000为A，列名设置为“价格等级”。

```python
hotel["价格等级"] = pd.cut(hotel["价格"],[0,500,1000,np.max(hotel["价格"])],labels = ["C","B","A"])
hotel
```

（7）获取评分均值最高和最低的地区的数据，分别使用append和concat方法将获取的两个数据集合并。

```python
dest_high = hotel["评分"].groupby(hotel["地区"]).mean().sort_values(ascending=False).index[0]
dest_low = hotel["评分"].groupby(hotel["地区"]).mean().sort_values().index[0]
score_high = hotel[hotel.地区 == dest_high]
score_low = hotel[hotel.地区 == dest_low]
```

```python
score_total = score_high.append(score_low)
score_total
```

```python
dff=pd.concat([score_high,score_low],axis=0) 
dff
```

（8）数据离散化，按照评分人数将酒店平均分为3个等级，三个等级的酒店数量尽量保持一致。评分人数最多的为A，最少的为C。列名设置为“热门等级”。

```python
hotel["热门等级"] = pd.qcut(hotel["评分人数"],q=3,labels=["C","B","A"])
hotel
```

（9）选出评分人数为A，价格也为A的酒店数据，计算其平均评分。

```python
hotel[(hotel.热门等级 == "A")&(hotel.价格等级 == "A")]
```

```python
hotel[(hotel.热门等级 == "A")&(hotel.价格等级 == "A")]["评分"].mean()
```

（11）纵向拆分数据集，分为df1和df2，df1包含名字，类型，城市，地区，df2包含名字，地点，评分，评分人数，价格，价格等级，热门等级。

```python
df1 = hotel[["名字","类型","城市","地区"]]
df1
df2 = hotel[["名字","地点","评分","评分人数","价格","价格等级","热门等级"]]
df2
```

（12）将df2按照价格进行排序，重新设置df2的索引。索引值等于价格排名。

```python
df2.copy().sort_values(by=["价格"],inplace=True)
df2.index = range(1,len(df2)+1)
df2
```

（13）使用merge方法将df1和df2合并。

```python
df_total = pd.merge(df1,df2,how='inner',on='名字')
```

（14）将合并后的数据集保存数据到“酒店数据2.xlsx”。

```python
df_total.to_excel("酒店数据2.xlsx")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['font.sans-serif']=['SimHei']  ##设置字体为 黑体
plt.rcParams['axes.unicode_minus']=False ##显示符号
```

（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
x = np.linspace(-5.0,3.0)
y = x**2+2*x+1
plt.plot(x,y,"b-")
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
x=np.linspace(-3.14,3.14)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,6))

plt.subplot(211)
plt.plot(x,y1,"r-")
plt.subplot(212)
plt.plot(x,y2,"b-")
plt.show()
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
df = pd.read_excel("酒店数据2.xlsx")
df.drop("Unnamed: 0",axis=1,inplace=True)
df
```

```python
data=df["地区"].value_counts()
```

```python
x=data.index
y=data.values
plt.figure(figsize=(10,6))
plt.bar(x,y,color="r")
plt.xticks(rotation=90)
plt.tick_params(labelsize=14)
plt.show()
```

（4）画出每个价格等级酒店数量的柱状图。

```python
data2=df["等级"].value_counts()
data2
```

```python
x=data2.index
y=data2.values
plt.figure(figsize=(8,6))
plt.bar(x,y,color="r")
plt.show()
```

（5）画出各个价格等级占比的饼图。

```python
plt.figure(figsize=(6,6))
plt.pie(y,labels=data2.index,autopct='%.1f %%',colors='byg',startangle=120)
plt.show()
```

（6）画出酒店评分的直方图。

```python
plt.figure(figsize=(10,6))
plt.hist(df['评分'],edgecolor='k',alpha=0.5)
plt.show()
```

（7）画出每个热门等级酒店评分均值的柱状图。（按照评分均值从小到大排序。）

```python
data3=df["评分"].groupby(df["等级"]).mean().sort_values()
```

```python
x=data3.index
y=data3.values
plt.figure(figsize=(4,6))
plt.bar(x,y,color="b")
plt.show()
```
